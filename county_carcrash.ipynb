```python
import pandas as pd
```

```python
raw_data = pd.read_csv("carcrash.csv", index_col="CRASH_CRN")
```

```python
raw_data.head(10)
```

```python
municipality = pd.read_csv("municipalitycodes.csv", index_col="Code")
```

```python
municipality.head(10)
```

```python
len(raw_data)
```

```python
raw_data.rename(columns = {"MUNICIPALITY":"Code"}, inplace = True)
crashes = raw_data.groupby("Code").count()
```

```python
crashes.head(10)
```

```python
crashes["CRASH_TOTAL"] = crashes["DISTRICT"]
```

```python
crashes = crashes.drop(crashes.columns.difference(["CRASH_TOTAL"]), axis=1)
```

```python
crashes.head(10)
```

```python
data = pd.merge(crashes, municipality, how="left", on="Code")
data.head(10)
```

```python
data = data.dropna()
```

```python
data = data.assign(Score = pd.cut(data["CRASH_TOTAL"], bins = [0, 200, 1000, 10000, 100000], labels = ["4", "3", "2", "1"]))
```

```python
data.sort_values(by=["CRASH_TOTAL"])
```
